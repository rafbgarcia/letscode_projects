## Investigate a melhor combinação de investimentos

```python
from collections import Counter
import random, itertools

def dist(vector1, vector2):
    if len(vector1) != len(vector2):
        raise ValueError('Feature vectors length must match')
    
    size = len(vector1)
    distances = [(vector1[i] - vector2[i]) ** 2 for i in range(size)]
    return sum(distances) ** 0.5

def wallets_dist(wallet1, wallet2, indexes):
    vec1 = [wallet1[i] for i in indexes]
    vec2 = [wallet2[i] for i in indexes]
    return dist(vec1, vec2)

def knn_prediction(clients, new_client, indexes, k):
    clients_sorted_by_nearest = sorted(clients, key=lambda client: wallets_dist(client.wallet_perc, new_client.wallet_perc, indexes))
    knn = clients_sorted_by_nearest[:k]
    counter = Counter(map(lambda client: client.profile, knn))
    return counter.most_common()[0][0]

def split_data(data):
    training_size = 70 / 100
    train_indexes = random.sample(range(len(data)), int(len(data) * training_size))
    test_indexes = [n for n in range(len(data)) if n not in train_indexes]
    training = [data[i] for i in train_indexes]
    test = [data[i] for i in test_indexes]
    return training, test
```

```python
feature_combinations = [list(itertools.combinations(range(4), n)) for n in range(1, 5)]
feature_combinations = [item for array in feature_combinations for item in array]

k = 11
combination_scores = {}
for indexes in feature_combinations:
    combination_scores[indexes] = []

    for _ in range(200):
        training, test = split_data(clients)
        right_predictions = 0
        for test_client in test:
            predicted_profile = knn_prediction(training, test_client, indexes, k)
            right_predictions += 1 if predicted_profile == test_client.profile else 0

        combination_scores[indexes].append(right_predictions / len(test))
```

```python
def get_mean(scores):
    return round(sum(scores) / len(scores), 3)

def get_std(mean, scores):
    squared_sum = sum([(mean - n) ** 2 for n in scores]) / (len(scores) - 1)
    return round(squared_sum ** 0.5, 3)

def get_ci(mean, std):
    lb = round(mean - 1.96 * std, 3)
    ub = round(mean + 1.96 * std, 3)
    ub = ub if ub < 1 else 1
    return (lb, ub)

sorted_scores = sorted(combination_scores.items(), key=lambda item: get_mean(item[1]))
combinations, means, stds, conf_intvs = [], [], [], []

for combination, scores in sorted_scores:
    combinations.append(', '.join(map(str, combination)))
    mean = get_mean(scores)
    std = get_std(mean, scores)
    means.append(mean)
    stds.append(std)
    conf_intvs.append(' / '.join(map(str, get_ci(mean, std))))
    
df = pd.DataFrame([['Mean'] + means, ['Std'] + stds, ['95%'] + conf_intvs], columns=[None] + combinations)
df = df.set_index(None)
df
```
